```python
import pickle
import os
import numpy as np
import pandas as pd
from matplotlib import cm
import matplotlib.pyplot as plt
plt.rcParams.update({'font.size': 18})
plt.rc('axes', axisbelow=True)

from utils.visualization import plot_median_initialization
```

```python
dir = '../../_data/fig-linear-self-attention/symmetric-dataset.pkl'

if os.path.isfile(dir):
    with open(dir, 'rb') as file:
        models = pickle.load(file)
else: models = {}

data_path = '../_data/custom-models/symmetric-attention-from-scratch/symmetric-attention-from-scratch-final/'
```

```python
## example encoder-mode with initialization
example = 'history_bert_small--encoder--red_pajama--symmetric-init_aqv6sj9j.csv'
data_encoder_init = pd.read_csv(data_path + example, 
                       encoding='UTF-8', 
                       delimiter=',')  

example = 'history_bert_small--encoder--red_pajama_tsfsgmxk.csv'
data_encoder = pd.read_csv(data_path + example, 
                       encoding='UTF-8', 
                       delimiter=',')  
```

```python
plt.figure(figsize=(5, 5))

layers = 4
heads = 4
plot_median_initialization(data_encoder_init, layers, heads, c = 'navy')
plot_median_initialization(data_encoder, layers, heads, c = 'lightblue')
# plt.plot(epoch, encoder_sym, color = 'lightblue')

plt.axhline(y = 0, linestyle = 'dashed', color = 'k')
plt.grid(color = 'gray', linestyle = '--', linewidth = 0.5)
plt.ylim(-1, 1)
plt.ylabel('score')
plt.xlabel('epochs')

plt.savefig(f'../figures/fig-symmetric-initialization-symmetry.pdf', format='pdf', dpi=300)
```

```python
plt.figure(figsize=(5, 5))

plt.plot(data_encoder_init['_step'], data_encoder_init['train/loss'], color = 'navy')
plt.plot(data_encoder['_step'], data_encoder['train/loss'], color = 'lightblue')

plt.axhline(y = 0, linestyle = 'dashed', color = 'k')
plt.grid(color = 'gray', linestyle = '--', linewidth = 0.5)
plt.yscale('log')
plt.ylabel('loss')
plt.xlabel('epochs')

plt.savefig(f'../figures/fig-symmetric-initialization-loss.pdf', format='pdf', dpi=300)
```

```python
def generate_color_list(n, cmap):
    colormap = cm.get_cmap(cmap, n)
    print(colormap(0))
    color_list = [colormap(i) for i in range(n)]
    return color_list

c = generate_color_list(layers, cmap = 'cividis')
```

```python
plt.figure(figsize=(5, 5))

layers = 4
heads = 4
plot_median_initialization(data_encoder_init, layers, heads, c = c, mode = 'layers')

# plt.legend()
plt.axhline(y=0,linestyle='dashed',color='k')
plt.grid(color = 'gray', linestyle = '--', linewidth = 0.5)
plt.ylim(-1, 1)
plt.ylabel('score')
plt.xlabel('epochs')

plt.savefig(f'../figures/fig-symmetric-initialization-symmetry-init-layers.pdf', format='pdf', dpi=300)
```

```python
plt.figure(figsize=(5, 5))

layers = 4
heads = 4
plot_median_initialization(data_encoder, layers, heads, c = c, mode = 'layers')

# plt.legend()
plt.axhline(y=0,linestyle='dashed',color='k')
plt.grid(color = 'gray', linestyle = '--', linewidth = 0.5)
plt.ylim(-1, 1)
plt.ylabel('score')
plt.xlabel('epochs')

plt.savefig(f'../figures/fig-symmetric-initialization-symmetry-layers.pdf', format='pdf', dpi=300)
```
