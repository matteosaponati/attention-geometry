```python
import pandas as pd
import matplotlib.pyplot as plt
plt.rcParams.update({'font.size': 18})
plt.rc('axes', axisbelow=True)
```

```python
data_path = '../_data/custom-models/sym-ssym-analysis-2/sym-ssym-analysis-2/'
```

```python
## example encoder-mode
example = 'vefpn7qr'
data = pd.read_csv(data_path + f'history_Default_{example}.csv', 
                       encoding='UTF-8', 
                       delimiter=',')  
encoder_sym = 2 * data['Median Symmetry'] - 1
encoder_accuracy = data['train/accuracy (missing token)']
epoch = data['epoch']

## example decoder-mode
example = 'e77889t2'
data = pd.read_csv(data_path + f'history_Default_{example}.csv', 
                       encoding='UTF-8', 
                       delimiter=',')  
decoder_sym = 2 * data['Median Symmetry'] - 1
decoder_accuracy = data['train/accuracy (missing token)']
epoch = data['epoch']
```

```python
plt.figure(figsize=(5, 5))

plt.plot(epoch, encoder_sym, color = 'navy')
plt.plot(epoch, decoder_sym, color = 'purple')

plt.axhline(y = 0, linestyle = 'dashed', color = 'k')
plt.grid(color = 'gray', linestyle = '--', linewidth = 0.5)
plt.ylim(-1, 1)
plt.ylabel('score')
plt.xlabel('epochs')

plt.savefig(f'../figures/fig-linear-self-attention-symmetric-dataset-symmetry.pdf', format='pdf', dpi=300)
```

```python
plt.figure(figsize=(5, 5))

plt.plot(epoch, encoder_accuracy, color = 'navy')
plt.plot(epoch, decoder_accuracy, color = 'purple')

plt.grid(color = 'gray', linestyle = '--', linewidth = 0.5)
plt.ylim(0, 1)
plt.ylabel('accuracy')
plt.xlabel('epochs')

plt.savefig(f'../figures/fig-linear-self-attention-symmetric-dataset-accuracy.pdf', format='pdf', dpi=300)
```
