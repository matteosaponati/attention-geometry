```python
import pickle
import numpy as np
import matplotlib.pyplot as plt
plt.rcParams.update({'font.size': 18})
plt.rc('axes', axisbelow=True)
from matplotlib import cm
from matplotlib.colors import LinearSegmentedColormap

# Create the colormap
colorList = ['purple', 'white', 'navy']
cmap = LinearSegmentedColormap.from_list('custom_diverging', colorList)
```

```python
def generate_color_list(n, cmap):
    colormap = cm.get_cmap(cmap, n)
    print(colormap(0))
    color_list = [colormap(i) for i in range(n)]
    return color_list
```

```python
with open('../_data/fig-symmetry-training-layers/BERT-base-encoder-specs.pkl', 'rb') as file:
    specs = pickle.load(file)
with open('../_data/fig-symmetry-training-layers/BERT-base-encoder-training.pkl', 'rb') as file:
    models = pickle.load(file)
```

```python
dataset = 'wikipedia'
```

```python
c = generate_color_list(models[f'bert-encoder-{dataset}'].shape[0], cmap = 'cividis')
plt.figure(figsize=(5, 5))

for l in range(models[f'bert-encoder-{dataset}'].shape[0]):

    scores = 2 * models[f'bert-encoder-{dataset}']  - 1
    plt.plot(specs[f'bert-encoder-{dataset}'][0], np.median(scores, axis = 1)[l, :], color = c[l], label = f'layer {l}')
    plt.fill_between(specs[f'bert-encoder-{dataset}'][0], 
                np.percentile(scores, 25, axis = 1)[l, :],
                np.percentile(scores, 75, axis = 1)[l, :],
                color = c[l], alpha = .3)

# plt.legend()
plt.axhline(y=0,linestyle='dashed',color='k')
plt.grid(color = 'gray', linestyle = '--', linewidth = 0.5)
plt.ylim(-1, 1)
plt.ylabel('score')
plt.xlabel('epochs')

plt.savefig(f'../figures/fig-bert-encoder-symmetry-layers-{dataset}.pdf', format='pdf', dpi=300)
```

```python
dataset = 'wikipedia'
```

```python
c = generate_color_list(models_encoder[f'bert-encoder-{dataset}'].shape[0], cmap = 'cividis')
plt.figure(figsize=(5, 5))

for l in range(models_encoder[f'bert-encoder-{dataset}'].shape[0]):

    scores = 2 * models_encoder[f'bert-encoder-{dataset}'][:, :, -1]  - 1
    median = np.median(scores, axis = 1)
    q1_range = median - np.percentile(scores, 25, axis = 1)
    q2_range = np.percentile(scores, 75, axis = 1) - median
    
    plt.errorbar(range(scores.shape[0]), median, yerr = [q1_range, q2_range], 
                     fmt = 'o', ecolor = 'darkblue', capsize = 5, marker = 'o', markersize = 5, color = 'darkblue')

    scores = 2 * models_decoder[f'bert-decoder-{dataset}'][:, :, -1]  - 1
    median = np.median(scores, axis = 1)
    q1_range = median - np.percentile(scores, 25, axis = 1)
    q2_range = np.percentile(scores, 75, axis = 1) - median
    
    plt.errorbar(range(scores.shape[0]), median, yerr = [q1_range, q2_range], 
                     fmt = 'o', ecolor = 'purple', capsize = 5, marker = 'o', markersize = 5, color = 'purple')

plt.axhline(y=0,linestyle='dashed',color='k')
plt.grid(color = 'gray', linestyle = '--', linewidth = 0.5)
plt.ylim(-1, 1)
plt.ylabel('score')
plt.xlabel('layers')

plt.savefig(f'../figures/fig-bert-symmetry-layers-{dataset}.pdf', format='pdf', dpi=300)
```
