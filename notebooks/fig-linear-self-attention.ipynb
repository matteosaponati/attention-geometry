```python
import pickle
import numpy as np
import matplotlib.pyplot as plt
plt.rcParams.update({'font.size': 18})
plt.rc('axes', axisbelow=True)
```

```python
with open('../_data/fig-linear-self-attention/linear-models.pkl', 'rb') as file:
    models = pickle.load(file)
```

```python
## example decoder-mode
decoder_sym = 2 * models[list(models.keys())[-9]][-2] - 1
decoder_accuracy = models[list(models.keys())[-9]][2]
epoch = models[list(models.keys())[-2]][-1]

## example encoder-mode
encoder_sym = 2 * models[list(models.keys())[-6]][6] - 1
encoder_accuracy = models[list(models.keys())[-6]][-4]
```

```python
plt.figure(figsize=(5, 5))

plt.plot(epoch, encoder_sym, color = 'navy')
plt.plot(epoch, decoder_sym, color = 'purple')
# plt.plot(2 * models['symmetric-loss'][1] - 1, color = 'navy')

plt.axhline(y = 0, linestyle = 'dashed', color = 'k')
plt.grid(color = 'gray', linestyle = '--', linewidth = 0.5)
plt.ylim(-1, 1)
plt.ylabel('score')
plt.xlabel('epochs')

plt.savefig(f'../figures/fig-linear-self-attention-symmetry.pdf', format='pdf', dpi=300)
```

```python
plt.figure(figsize=(5, 5))

plt.plot(epoch, decoder_accuracy, color = 'purple')
plt.plot(epoch, encoder_accuracy, color = 'navy')

plt.grid(color = 'gray', linestyle = '--', linewidth = 0.5)
plt.ylim(0, 1)
plt.ylabel('accuracy')
plt.xlabel('epochs')

plt.savefig(f'../figures/fig-linear-self-attention-accuracy.pdf', format='pdf', dpi=300)
```
