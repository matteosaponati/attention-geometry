```python
import pickle
import matplotlib.pyplot as plt
plt.rcParams.update({'font.size': 22})
plt.rc('axes', axisbelow=True)

from utils.visualization import plot_scores
```

```python
with open('../_results/custom-models/bert-encoder-jigsaw.pkl', 'rb') as file:
    encoder_jigsaw = pickle.load(file)
with open('../_results/custom-models/bert-encoder-red_pajama.pkl', 'rb') as file:
    encoder_red_pajama = pickle.load(file)
with open('../_results/custom-models/bert-decoder-jigsaw.pkl', 'rb') as file:
    decoder_jigsaw = pickle.load(file)
with open('../_results/custom-models/bert-decoder-red_pajama.pkl', 'rb') as file:
    decoder_red_pajama = pickle.load(file)
```

```python
plt.figure(figsize=(5,5))

plot_scores(encoder_jigsaw, color = 'navy')
plot_scores(decoder_jigsaw, color = 'purple')

plt.axhline(y=0,linestyle='dashed',color='k')
plt.grid(color = 'gray', linestyle = '--', linewidth = 0.5)
plt.ylim(-1,1)
plt.ylabel('symmetry score')
plt.xticks([0, 100, 200], [0, 1000, 2000])
plt.xlabel('epochs')

plt.savefig('figures/suppfig-symmetry-custom-jigsaw.pdf', format='pdf', dpi=300)
```

```python
plt.figure(figsize=(5,5))

plot_scores(encoder_jigsaw, color = 'navy', mode = 'directionality')
plot_scores(decoder_jigsaw, color = 'purple', mode = 'directionality')

plt.axhline(y=0,linestyle='dashed',color='k')
plt.grid(color = 'gray', linestyle = '--', linewidth = 0.5)
plt.ylim(-1,1)
plt.ylabel('directionality score')
plt.xticks([0, 100, 200], [0, 1000, 2000])
plt.xlabel('epochs')

plt.savefig('figures/suppfig-directionality-custom-jigsaw.pdf', format='pdf', dpi=300)
```

```python
plt.figure(figsize=(5,5))

plot_scores(encoder_red_pajama, color = 'navy')
plot_scores(decoder_red_pajama, color = 'purple')

plt.axhline(y=0,linestyle='dashed',color='k')
plt.grid(color = 'gray', linestyle = '--', linewidth = 0.5)
plt.ylim(-1,1)
plt.ylabel('symmetry score')
plt.xticks([0, 100, 200], [0, 1000, 2000])
plt.xlabel('epochs')

plt.savefig('figures/suppfig-symmetry-custom-red-pajama.pdf', format='pdf', dpi=300)
```
