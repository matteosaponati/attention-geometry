```python
import pickle
import matplotlib.pyplot as plt
plt.rcParams.update({'font.size': 22})
plt.rc('axes', axisbelow=True)

from utils.visualization import plot_scores_scatter
```

```python
with open('../_results/vision-models/BEiT.pkl', 'rb') as file:
    BEiT = pickle.load(file)
with open('../_results/vision-models/DINO.pkl', 'rb') as file:
    DINO = pickle.load(file)
with open('../_results/vision-models/DIT.pkl', 'rb') as file:
    DIT = pickle.load(file)
with open('../_results/vision-models/GIT-images.pkl', 'rb') as file:
    GIT_images = pickle.load(file)
with open('../_results/vision-models/DEIT.pkl', 'rb') as file:
    DEIT = pickle.load(file)
with open('../_results/vision-models/DETR-Encoder.pkl', 'rb') as file:
    DETR_enc = pickle.load(file)
with open('../_results/vision-models/DINOv2.pkl', 'rb') as file:
    DINOv2 = pickle.load(file)
with open('../_results/vision-models/ViT-21k.pkl', 'rb') as file:
    vit_21k = pickle.load(file)
with open('../_results/vision-models/ViT.pkl', 'rb') as file:
    vit = pickle.load(file)
with open('../_results/vision-models/ViT-MAE.pkl', 'rb') as file:
    vit_mae = pickle.load(file)

with open('../_results/vision-models/iGPT.pkl', 'rb') as file:
    iGPT = pickle.load(file)
with open('../_results/vision-models/DETR-Decoder.pkl', 'rb') as file:
    DETR_dec = pickle.load(file)
with open('../_results/vision-models/mask2former-decoder.pkl', 'rb') as file:
    m2f_decoder = pickle.load(file)
with open('../_results/vision-models/LLAMA.pkl', 'rb') as file:
    LLAMA = pickle.load(file)
```

```python
plt.figure(figsize=(5,5))

plot_scores_scatter(BEiT, color = 'navy')
plot_scores_scatter(DINO, color = 'navy')
plot_scores_scatter(DIT, color = 'navy')
plot_scores_scatter(GIT_images, color = 'navy')
plot_scores_scatter(DEIT, color = 'navy')
plot_scores_scatter(DETR_enc, color = 'navy')
plot_scores_scatter(DINOv2, color = 'navy')
plot_scores_scatter(vit_21k, color = 'navy')
plot_scores_scatter(vit, color = 'navy')
plot_scores_scatter(vit_mae, color = 'navy')

plot_scores_scatter(iGPT, color = 'purple')
plot_scores_scatter(DETR_dec, color = 'purple')
plot_scores_scatter(m2f_decoder, color = 'purple')
plot_scores_scatter(LLAMA, color = 'purple')

plt.axhline(y=0,linestyle='dashed',color='k')
plt.grid(color = 'gray', linestyle = '--', linewidth = 0.5)
plt.xscale('log')
plt.ylim(-1,1)
plt.xlim(10e5, 10e11)
plt.ylabel('symmetry score')
plt.xlabel('# parameters')

plt.savefig('figures/fig-symmetry-vision.pdf', format='pdf', dpi=300)
```
