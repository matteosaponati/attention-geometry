```python
import pickle
import matplotlib.pyplot as plt
plt.rcParams.update({'font.size': 22})
plt.rc('axes', axisbelow=True)

from utils.visualization import plot_median_scores
```

```python
with open('../_results/language-models/BERT.pkl', 'rb') as file:
    BERT = pickle.load(file)
with open('../_results/language-models/ALBERT.pkl', 'rb') as file:
    ALBERT = pickle.load(file)
with open('../_results/language-models/ROBERTA.pkl', 'rb') as file:
    ROBERTA = pickle.load(file)
with open('../_results/language-models/DEBERTA.pkl', 'rb') as file:
    DEBERTA = pickle.load(file)
with open('../_results/language-models/ModernBERT.pkl', 'rb') as file:
    ModernBERT = pickle.load(file)

with open('../_results/language-models/GPT.pkl', 'rb') as file:
    GPT = pickle.load(file)
with open('../_results/language-models/GPT-neo.pkl', 'rb') as file:
    GPTneo = pickle.load(file)
with open('../_results/language-models/Tiny-GPT.pkl', 'rb') as file:
    TinyGPT = pickle.load(file)
with open('../_results/language-models/LLAMA.pkl', 'rb') as file:
    LLAMA = pickle.load(file)
with open('../_results/language-models/MISTRAL.pkl', 'rb') as file:
    MISTRAL = pickle.load(file)
with open('../_results/language-models/MobileLLM.pkl', 'rb') as file:
    MobileLLM = pickle.load(file)
with open('../_results/language-models/phi-1.pkl', 'rb') as file:
    phi_1 = pickle.load(file)
with open('../_results/language-models/phi-2.pkl', 'rb') as file:
    phi_2 = pickle.load(file)
with open('../_results/language-models/phi-3.pkl', 'rb') as file:
    phi_3 = pickle.load(file)
```

```python
plt.figure(figsize=(5,5))

plot_median_scores(BERT, color = 'navy')
plot_median_scores(ALBERT, color = 'navy')
plot_median_scores(ROBERTA, color = 'navy')
plot_median_scores(DEBERTA, color = 'navy')
plot_median_scores(ModernBERT, color = 'navy')

plot_median_scores(GPT, color = 'purple')
plot_median_scores(GPTneo, color = 'purple')
plot_median_scores(TinyGPT, color = 'purple')
plot_median_scores(MISTRAL, color = 'purple')
plot_median_scores(LLAMA, color = 'purple')
plot_median_scores(MobileLLM, color = 'purple')
plot_median_scores(phi_1, color = 'purple')
plot_median_scores(phi_2, color = 'purple')
plot_median_scores(phi_3, color = 'purple')

plt.axhline(y=0,linestyle='dashed',color='k')
plt.grid(color = 'gray', linestyle = '--', linewidth = 0.5)
plt.xscale('log')
plt.ylim(-1,1)
plt.xlim(10e5, 10e11)
plt.ylabel('symmetry score')
plt.xlabel('# parameters')

plt.savefig('figures/fig-symmetry-language.pdf', format='pdf', dpi=300)
```
