```python
import pickle
import matplotlib.pyplot as plt
plt.rcParams.update({'font.size': 18})
plt.rc('axes', axisbelow=True)

from utils.visualization import plot_median_scores
```

```python
with open('../_results/language-models/T5Encoder.pkl', 'rb') as file:
    T5Encoder = pickle.load(file)
with open('../_results/language-models/BARTEncoder.pkl', 'rb') as file:
    BARTEncoder = pickle.load(file)
with open('../_results/language-models/T5Decoder.pkl', 'rb') as file:
    T5Decoder = pickle.load(file)
with open('../_results/language-models/BARTDecoder.pkl', 'rb') as file:
    BARTDecoder = pickle.load(file)
```

```python
plt.figure(figsize=(5,5))

plot_median_scores(T5Encoder, color = 'navy')
plot_median_scores(BARTEncoder, color = 'navy')

plot_median_scores(T5Decoder, color = 'purple')
plot_median_scores(BARTDecoder, color = 'purple')

plt.axhline(y=0,linestyle='dashed',color='k')
plt.grid(color = 'gray', linestyle = '--', linewidth = 0.5)
plt.xscale('log')
plt.ylim(-1,1)
plt.xlim(10e5, 10e11)
plt.ylabel('symmetry score')
plt.xlabel('# parameters')

plt.savefig('figures/suppfig-symmetry-encoder-decoder.pdf', format='pdf', dpi=300)
```

```python
plt.figure(figsize=(5,5))

plot_median_scores(T5Encoder, color = 'navy', mode = 'directionality')
plot_median_scores(BARTEncoder, color = 'navy', mode = 'directionality')

plot_median_scores(T5Decoder, color = 'purple', mode = 'directionality')
plot_median_scores(BARTDecoder, color = 'purple', mode = 'directionality')

plt.axhline(y=0,linestyle='dashed',color='k')
plt.grid(color = 'gray', linestyle = '--', linewidth = 0.5)
plt.xscale('log')
plt.ylim(-1,1)
plt.xlim(10e5, 10e11)
plt.ylabel('directionality score')
plt.xlabel('# parameters')

plt.savefig('figures/suppfig-directionality-encoder-decoder.pdf', format='pdf', dpi=300)
```
