```python
import pickle
import matplotlib.pyplot as plt
plt.rcParams.update({'font.size': 18})
plt.rc('axes', axisbelow=True)

from utils.visualization import plot_scores_layers
```

```python
with open('../_results/custom-models/bert-encoder-wikipedia.pkl', 'rb') as file:
    encoder_wikipedia = pickle.load(file)
with open('../_results/custom-models/bert-decoder-wikipedia.pkl', 'rb') as file:
    decoder_wikipedia = pickle.load(file)
```

```python
plt.figure(figsize=(5,5))

plot_scores_layers(encoder_wikipedia)

plt.axhline(y=0,linestyle='dashed',color='k')
plt.grid(color = 'gray', linestyle = '--', linewidth = 0.5)
plt.ylim(-1,1)
plt.ylabel('symmetry score')
plt.xticks([0, 100, 200], [0, 1000, 2000])
plt.xlabel('epochs')

plt.savefig('figures/fig-symmetry-layers-custom-encoder.pdf', format='pdf', dpi=300)
```

```python
plt.figure(figsize=(5,5))

plot_scores_layers(decoder_wikipedia)

plt.axhline(y=0,linestyle='dashed',color='k')
plt.grid(color = 'gray', linestyle = '--', linewidth = 0.5)
plt.ylim(-1,1)
plt.ylabel('symmetry score')
plt.xticks([0, 100, 200], [0, 1000, 2000])
plt.xlabel('epochs')

plt.savefig('figures/fig-symmetry-layers-custom-decoder.pdf', format='pdf', dpi=300)
```

```python
plt.figure(figsize=(5,5))

plot_scores_layers(encoder_wikipedia, mode = 'directionality')

plt.axhline(y=0,linestyle='dashed',color='k')
plt.grid(color = 'gray', linestyle = '--', linewidth = 0.5)
plt.ylim(-1,1)
plt.ylabel('directionality score')
plt.xticks([0, 100, 200], [0, 1000, 2000])
plt.xlabel('epochs')

plt.savefig('figures/fig-directionality-layers-custom-encoder.pdf', format='pdf', dpi=300)
```
