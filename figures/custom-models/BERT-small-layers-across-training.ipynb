```python
import pickle
import numpy as np
from scipy.stats import sem
import matplotlib.pyplot as plt
```

```python
# bert-small 
with open('../../_data/pascal-data/symmetry-scores/BERT-small-symmetry.pkl', 'rb') as file:
    models = pickle.load(file)
```

```python
def get_averages_across_layers(model, h, l):

    meanList = np.zeros((model[0].size, l))
    stdList = np.zeros((model[0].size, l))

    for i in range(l):

        collect = np.zeros((model[0].size, h))

        for j in range(h):

            collect[:, j] = model[7 + i*h + j]

        meanList[:, i] = collect.mean(axis = 1)
        stdList[:, i] = collect.std(axis = 1)

    return meanList, stdList
```

```python
model = models['bert-small-encoder-jigsaw']

mean, std = get_averages_across_layers(model, h = 4, l = 4)
c = ['blue', 'slateblue', 'purple', 'mediumvioletred']

plt.figure(figsize=(4,4))

for i in range(4):
    
    plt.plot(model[0], mean[:, i], 
         color = c[i], label = f'layer {i}')
    plt.fill_between(model[0], mean[:, i] - std[:, i],
                 mean[:, i] + std[:, i],
                 color = c[i], alpha = .3)
    
plt.grid(color = 'gray', linestyle = '--', linewidth = 0.5)
plt.ylabel('scores')
plt.xlabel('epochs')
plt.title('encoder-only (4 l, 4 h)')
plt.ylim(0,1)

plt.axhline(y = .5,color='k')
plt.legend()
    
```
