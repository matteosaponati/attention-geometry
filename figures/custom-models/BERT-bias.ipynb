```python
import pickle
import numpy as np
from scipy.stats import sem
import matplotlib.pyplot as plt
```

```python
## BERT base reference

## BERT base reference
with open('../../_data/fig_scores/BERT-query-key.pkl', 'rb') as file:
    models = pickle.load(file)

BERT_small_mean = models['BERTbase'][-2].mean()
```

```python
plt.figure(figsize=(5,4))

# bert-small 
with open('../../_data/pascal-data/symmetry-scores/BERT-symmetry.pkl', 'rb') as file:
    models = pickle.load(file)

step = models['bert-encoder-jigsaw'][0]
symmetry_scores = models['bert-encoder-jigsaw'][2]
var_scores = models['bert-encoder-jigsaw'][6]
plt.plot(step, symmetry_scores, 
         color = 'darkblue', linestyle = 'dashed')
plt.fill_between(step, symmetry_scores - var_scores,
                 symmetry_scores + var_scores,
                 color = 'darkblue', alpha = .3)

step = models['bert-encoder-wikipedia'][0]
symmetry_scores = models['bert-encoder-wikipedia'][2]
var_scores = models['bert-encoder-wikipedia'][6]
plt.plot(step, symmetry_scores, 
         color = 'blue', linestyle = 'dashed')
plt.fill_between(step, symmetry_scores - var_scores,
                 symmetry_scores + var_scores,
                 color = 'blue', alpha = .3)

step = models['bert-encoder-red_pajama'][0]
symmetry_scores = models['bert-encoder-red_pajama'][2]
var_scores = models['bert-encoder-red_pajama'][6]
plt.plot(step, symmetry_scores, 
         color = 'slateblue', linestyle = 'dashed')
plt.fill_between(step, symmetry_scores - var_scores,
                 symmetry_scores + var_scores,
                 color = 'slateblue', alpha = .3)

step = models['bert-no_bias-encoder-jigsaw'][0]
symmetry_scores = models['bert-no_bias-encoder-jigsaw'][2]
var_scores = models['bert-no_bias-encoder-jigsaw'][6]
plt.plot(step, symmetry_scores, 
         color = 'darkblue')
plt.fill_between(step, symmetry_scores - var_scores,
                 symmetry_scores + var_scores,
                 color = 'darkblue', alpha = .3)

step = models['bert-no_bias-encoder-wikipedia'][0]
symmetry_scores = models['bert-no_bias-encoder-wikipedia'][2]
var_scores = models['bert-no_bias-encoder-wikipedia'][6]
plt.plot(step, symmetry_scores, 
         color = 'blue')
plt.fill_between(step, symmetry_scores - var_scores,
                 symmetry_scores + var_scores,
                 color = 'blue', alpha = .3)

step = models['bert-no_bias-encoder-red_pajama'][0]
symmetry_scores = models['bert-no_bias-encoder-red_pajama'][2]
var_scores = models['bert-no_bias-encoder-red_pajama'][6]
plt.plot(step, symmetry_scores, 
         color = 'slateblue')
plt.fill_between(step, symmetry_scores - var_scores,
                 symmetry_scores + var_scores,
                 color = 'slateblue', alpha = .3)

plt.grid(color = 'gray', linestyle = '--', linewidth = 0.5)
plt.ylabel('scores')
plt.xlabel('epochs')
plt.title('bert-like models (12 l, 12 h)')
plt.ylim(0,1)

plt.axhline(y = .5,color='k')
plt.axhline(y = BERT_small_mean,color='lightblue',linestyle='dashed', label = 'bert-mean')

plt.legend()
```
