```python
import pickle
import numpy as np
from scipy.stats import sem
import matplotlib.pyplot as plt
from matplotlib import cm
```

```python
# bert-small 
with open('../../_data/pascal-data/symmetry-scores/BERT-symmetry.pkl', 'rb') as file:
    models = pickle.load(file)

h = 12
l = 12
```

```python
def get_averages_across_layers(model, h, l):

    meanList = np.zeros((model[0].size, l))
    stdList = np.zeros((model[0].size, l))

    for i in range(l):

        collect = np.zeros((model[0].size, h))

        for j in range(h):

            collect[:, j] = model[7 + i*h + j]

        meanList[:, i] = collect.mean(axis = 1)
        stdList[:, i] = collect.std(axis = 1)

    return meanList, stdList
```

```python
def generate_color_list(n):
    # Generate a list of N colors from a colormap
    colormap = cm.get_cmap('inferno', n)
    color_list = [colormap(i) for i in range(n)]
    return color_list
```

```python
model = models['bert-encoder-jigsaw']

mean, std = get_averages_across_layers(model, h = h, l = l)

c = generate_color_list(l)

plt.figure(figsize=(4,4))

for i in range(l):
    
    plt.plot(model[0], mean[:, i], 
         color = c[i], label = f'layer {i}')
    plt.fill_between(model[0], mean[:, i] - std[:, i],
                 mean[:, i] + std[:, i],
                 color = c[i], alpha = .3)
    
plt.grid(color = 'gray', linestyle = '--', linewidth = 0.5)
plt.ylabel('scores')
plt.xlabel('epochs')
plt.title('encoder-only (12 l, 12 h)')
plt.ylim(0,1)

plt.axhline(y = .5,color='k')
plt.legend()
```
