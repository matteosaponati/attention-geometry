```python
import pickle
import numpy as np
from scipy.stats import sem
import matplotlib.pyplot as plt
```

```python
"""
Visualization of the symmetry and anti-symmetry scores across layers and heads
"""

with open('../data/fig_scores/models.pkl', 'rb') as file:
    models = pickle.load(file)

SList = [models[key][-2].flatten() for key in list(models.keys())]
NList = [models[key][-1].flatten() for key in list(models.keys())]
names = list(models.keys())

plt.figure(figsize=(10,4))

for i, model in enumerate(SList):
    
    plt.scatter(i*np.ones(SList[i].size),(SList[i].flatten()**2),color='purple',alpha=.01)
    plt.scatter(i,(SList[i].flatten()**2).mean(),color='purple')

    plt.scatter((i+.1)*np.ones(NList[i].size),(NList[i].flatten()**2),color='darkred',alpha=.01)
    plt.scatter(i+.1,(NList[i].flatten()**2).mean(),color='darkred')
    
plt.grid(color = 'gray', linestyle = '--', linewidth = 0.5)
plt.title('relative weight of S and N matrices')
plt.ylabel('scores')
plt.xticks(np.arange(len(names)),names)
plt.xticks(rotation=70)
plt.ylim(0,1)
```

```python
"""
Visualization of the symmetry and anti-symmetry scores across layers and heads
"""

with open('../data/fig_scores/models.pkl', 'rb') as file:
    models = pickle.load(file)

SList = [models[key][-2].flatten() for key in list(models.keys())]
NList = [models[key][-1].flatten() for key in list(models.keys())]
names = list(models.keys())

# Create positions for each violin plot
positions_s = [i for i in range(len(SList))]
positions_n = [i + 0.1 for i in range(len(NList))] 

plt.figure(figsize=(16,4))


# Plot violin plots for SList
plt.violinplot(SList, positions=positions_s, widths=.5, showmeans=True, showextrema=True, showmedians=True)#, color='purple')

# Plot violin plots for NList
plt.violinplot(NList, positions=positions_n, widths=.5, showmeans=True, showextrema=True, showmedians=True)#, color='darkred')
    
plt.grid(color = 'gray', linestyle = '--', linewidth = 0.5)
plt.title('relative weight of S and N matrices')
plt.ylabel('scores')
plt.xticks(np.arange(len(names)),names)
plt.xticks(rotation=70)
plt.ylim(0,1)
```
