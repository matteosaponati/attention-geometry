```python
import pickle
import numpy as np
from scipy.stats import sem
import matplotlib.pyplot as plt
```

```python
"""
Visualization of level of symmetry between different models 
(across layers and heads)
"""

# get scores for BERT models calculated on each head
with open('../../_data/fig_scores/BERT-query-key.pkl', 'rb') as file:
    models = pickle.load(file)

SList = [models[key][-2].flatten() for key in list(models.keys())]
names = list(models.keys())
```

```python
plt.figure(figsize=(8,4))

## BERT models
plt.boxplot(SList)

plt.grid(color = 'gray', linestyle = '--', linewidth = 0.5)
plt.title('BERT models')
plt.ylabel('scores')
plt.xticks(np.arange(1,len(names)+1),names)
plt.xticks(rotation=50)
plt.ylim(0,1)
plt.axhline(y=.5,color='k',linestyle='dashed')
```
