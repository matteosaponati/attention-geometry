```python
import pickle
import numpy as np
import matplotlib.pyplot as plt
```

```python
'upload scores (across layers) for all models'

## BERT models
with open('../../_data/fig_scores/BERT-query-key-full.pkl', 'rb') as file:
    models = pickle.load(file)

BERT_hist = []
for key in list(models.keys()):
    for k in models[key][-2].flatten().flatten()**2:
        BERT_hist.append(k)

## GPT models
with open('../../_data/fig_scores/GPT-query-key-full.pkl', 'rb') as file:
    models = pickle.load(file)

GPT_hist = []
for key in list(models.keys()):
    for k in models[key][-2].flatten().flatten()**2:
        GPT_hist.append(k)
```

```python
plt.figure(figsize=(5,4))

plt.hist(BERT_hist,weights = np.ones_like(BERT_hist) / len(BERT_hist),bins=30,color='darkblue',alpha=.3,label='BERT')
plt.axvline(x = np.mean(BERT_hist),linestyle='dashed',color='darkblue')
plt.hist(GPT_hist,weights = np.ones_like(GPT_hist) / len(GPT_hist),bins=30,color='palevioletred',alpha=.3,label='GPT')
plt.axvline(x = np.mean(GPT_hist),linestyle='dashed',color='palevioletred')

plt.xlim(0,1)
plt.axvline(x=.5,linestyle='dashed',color='k')
plt.ylim(0,.3)
plt.legend()
plt.xlabel('symmetry score')
plt.ylabel('occurrence (norm)')
plt.title('full matrix')
plt.grid(color = 'gray', linestyle = '--', linewidth = 0.5)
```
