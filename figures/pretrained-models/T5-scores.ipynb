```python
import pickle
import numpy as np
from scipy.stats import sem
import matplotlib.pyplot as plt
```

```python
"""
Visualization of level of symmetry between different models 
(across layers and heads)
"""

plt.figure(figsize=(8,4))

## T5Encoder models
plt.subplot(1,2,1)

# get scores for T5 Encoder models calculated on each head
with open('../../_data/fig_scores/T5Encoder-query-key.pkl', 'rb') as file:
    models = pickle.load(file)

SList = [models[key][-2].flatten()**2 for key in list(models.keys())]
names = list(models.keys())

for i, model in enumerate(SList):
    
    plt.scatter(i*np.ones(SList[i].size),SList[i],color='plum',alpha=.01)
    plt.scatter(i,(SList[i].flatten()).mean(),color='plum')

# get scores for T5 Encoder models calculated on the whole M matrix
with open('../../_data/fig_scores/T5Encoder-query-key-full.pkl', 'rb') as file:
    models = pickle.load(file)

SList = [models[key][-2].flatten()**2 for key in list(models.keys())]
names = list(models.keys())

for i, model in enumerate(SList):
    
    plt.scatter(i*np.ones(SList[i].size),SList[i],color='plum',alpha=.1,facecolors='none')
    plt.scatter(i,(SList[i].flatten()).mean(),color='plum',facecolors='none')

plt.grid(color = 'gray', linestyle = '--', linewidth = 0.5)
plt.title('T5 (encoder) models')
plt.ylabel('scores')
plt.xticks(np.arange(len(names)),names)
plt.xticks(rotation=50)
plt.ylim(0,1)
plt.axhline(y=.5,color='k',linestyle='dashed')

## T5 Decoder
plt.subplot(1,2,2)

with open('../../_data/fig_scores/T5Decoder-query-key.pkl', 'rb') as file:
    models = pickle.load(file)

SList = [models[key][-2].flatten()**2 for key in list(models.keys())]
names = list(models.keys())

for i, model in enumerate(SList):
    
    plt.scatter(i*np.ones(SList[i].size),SList[i],color='plum',alpha=.01)
    plt.scatter(i,(SList[i].flatten()).mean(),color='plum')

# get scores for T5Encoder models calculated on the whole M matrix
with open('../../_data/fig_scores/T5Decoder-query-key-full.pkl', 'rb') as file:
    models = pickle.load(file)

SList = [models[key][-2].flatten()**2 for key in list(models.keys())]
names = list(models.keys())

for i, model in enumerate(SList):
    
    plt.scatter(i*np.ones(SList[i].size),SList[i],color='plum',alpha=.01,facecolors='none')
    plt.scatter(i,(SList[i].flatten()).mean(),color='plum',facecolors='none')

plt.grid(color = 'gray', linestyle = '--', linewidth = 0.5)
plt.title('T5 (decoder) models')
plt.xticks(np.arange(len(names)),names)
plt.xticks(rotation=50)
plt.ylim(0,1)
plt.axhline(y=.5,color='k',linestyle='dashed')
```
