```python
import pickle
import numpy as np
from scipy.stats import sem
import matplotlib.pyplot as plt
```

```python
"""
Visualization of level of symmetry between different models 
(across layers and heads)
"""

plt.figure(figsize=(12,4))

## BERT models
plt.subplot(1,3,1)

# get scores for BERT models calculated on each head
with open('../../_data/fig_scores/BERT-query-key.pkl', 'rb') as file:
    models = pickle.load(file)

SList = [models[key][-2].flatten() for key in list(models.keys())]
names = list(models.keys())

plt.boxplot(SList)

plt.grid(color = 'gray', linestyle = '--', linewidth = 0.5)
plt.title('BERT models')
plt.ylabel('scores')
plt.xticks(np.arange(1,len(names)+1),names)
plt.xticks(rotation=50)
plt.ylim(0,1)
plt.axhline(y=.5,color='k',linestyle='dashed')

## ROBERTA
plt.subplot(1,3,2)

with open('../../_data/fig_scores/ROBERTA-query-key.pkl', 'rb') as file:
    models = pickle.load(file)

SList = [models[key][-2].flatten()**2 for key in list(models.keys())]
names = list(models.keys())

plt.boxplot(SList)

plt.grid(color = 'gray', linestyle = '--', linewidth = 0.5)
plt.title('ROBERTA models')
plt.ylabel('scores')
plt.xticks(np.arange(1,len(names)+1),names)
plt.xticks(rotation=50)
plt.ylim(0,1)
plt.axhline(y=.5,color='k',linestyle='dashed')

## ALBERT
plt.subplot(1,3,3)

with open('../../_data/fig_scores/ALBERT-query-key.pkl', 'rb') as file:
    models = pickle.load(file)
    
SList = [models[key][-2].flatten()**2 for key in list(models.keys())]
names = list(models.keys())

plt.boxplot(SList)

plt.grid(color = 'gray', linestyle = '--', linewidth = 0.5)
plt.title('ALBERT models')
plt.ylabel('scores')
plt.xticks(np.arange(1,len(names)+1),names)
plt.xticks(rotation=50)
plt.ylim(0,1)
plt.axhline(y=.5,color='k',linestyle='dashed')
```
