```python
import pickle
import numpy as np
from scipy.stats import sem
import matplotlib.pyplot as plt
```

```python
# get scores for BERT-base 
with open('../data/fig_scores/BERTmodels.pkl', 'rb') as file:
    models = pickle.load(file)
SBERTbase = models['BERTbase'][-2]**2
SBERTlarge = models['BERTlarge'][-2]**2

with open('../data/fig_scores/BERTFullmodels.pkl', 'rb') as file:
    models = pickle.load(file)
SBERTbaseFull = models['BERTbase'][-2]**2
SBERTlargeFull = models['BERTlarge'][-2]**2

# get scores for GPT-1 
with open('../data/fig_scores/GPTmodels.pkl', 'rb') as file:
    models = pickle.load(file)
SGPT1 = models['GPT'][-2]**2
SGPT2 = models['GPT2'][-2]**2

with open('../data/fig_scores/GPTFullmodels.pkl', 'rb') as file:
    models = pickle.load(file)
SGPT1Full = models['GPT'][-2]**2
SGPT2Full = models['GPT2'][-2]**2
```

```python
plt.figure(figsize=(10,4))

plt.subplot(2,2,1)

plt.scatter(np.arange(SGPT1Full.shape[0]),SGPT1Full,color='darkblue',marker='*')

for layer in range(SGPT1.shape[0]):
    
    plt.scatter(layer*np.ones(SGPT1.shape[1]),SGPT1[layer,:],color='darkblue',alpha=.1)
    plt.scatter(layer,SGPT1[layer,:].mean(),color='darkblue')

plt.grid(color = 'gray', linestyle = '--', linewidth = 0.5)
plt.title('GPT-1')
plt.ylabel('scores')
plt.xlabel('layers')
plt.ylim(0,1)
plt.axhline(y=.5,color='k',linestyle='dashed')

plt.subplot(2,2,2)

plt.scatter(np.arange(SBERTbaseFull.shape[0]),SBERTbaseFull,color='purple',marker='*')

for layer in range(SBERTbase.shape[0]):
    
    plt.scatter(layer*np.ones(SBERTbase.shape[1]),SBERTbase[layer,:],color='purple',alpha=.1)
    plt.scatter(layer,SBERTbase[layer,:].mean(),color='purple')

plt.grid(color = 'gray', linestyle = '--', linewidth = 0.5)
plt.title('BERT-base')
plt.ylabel('scores')
plt.xlabel('layers')
plt.ylim(0,1)
plt.axhline(y=.5,color='k',linestyle='dashed')

plt.subplot(2,2,3)

plt.scatter(np.arange(SGPT2Full.shape[0]),SGPT2Full,color='darkblue',marker='*')

for layer in range(SGPT2.shape[0]):
    
    plt.scatter(layer*np.ones(SGPT2.shape[1]),SGPT2[layer,:],color='darkblue',alpha=.1)
    plt.scatter(layer,SGPT2[layer,:].mean(),color='darkblue')

plt.grid(color = 'gray', linestyle = '--', linewidth = 0.5)
plt.title('GPT-2')
plt.ylabel('scores')
plt.xlabel('layers')
plt.ylim(0,1)
plt.axhline(y=.5,color='k',linestyle='dashed')

plt.subplot(2,2,4)

plt.scatter(np.arange(SBERTlargeFull.shape[0]),SBERTlargeFull,color='purple',marker='*')

for layer in range(SBERTlarge.shape[0]):
    
    plt.scatter(layer*np.ones(SBERTlarge.shape[1]),SBERTlarge[layer,:],color='purple',alpha=.1)
    plt.scatter(layer,SBERTlarge[layer,:].mean(),color='purple')

plt.grid(color = 'gray', linestyle = '--', linewidth = 0.5)
plt.title('BERT-base')
plt.ylabel('scores')
plt.xlabel('layers')
plt.ylim(0,1)
plt.axhline(y=.5,color='k',linestyle='dashed')
```
