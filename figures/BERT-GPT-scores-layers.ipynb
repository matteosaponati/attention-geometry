```python
import pickle
import numpy as np
from scipy.stats import sem, iqr
import matplotlib.pyplot as plt
```

```python
with open('../_data/fig_scores/BERT-query-key.pkl', 'rb') as file:
    models = pickle.load(file)

BERTbase = models['BERTbase'][-2]**2

with open('../_data/fig_scores/BERT-query-key-full.pkl', 'rb') as file:
    models = pickle.load(file)

BERTbase_full = models['BERTbase'][-2]**2

with open('../_data/fig_scores/GPT-query-key.pkl', 'rb') as file:
    models = pickle.load(file)

GPT1 = models['GPT'][-2]**2

with open('../_data/fig_scores/GPT-query-key-full.pkl', 'rb') as file:
    models = pickle.load(file)

GPT1_full = models['GPT'][-2]**2
```

```python
plt.figure(figsize=(10,4))

plt.subplot(1,2,1)

plt.scatter(np.arange(BERTbase_full.shape[0]),BERTbase_full,color='black')

plt.plot(np.arange(BERTbase.shape[0]),np.median(BERTbase,axis=1),color='purple')
plt.fill_between(np.arange(BERTbase.shape[0]),
                 np.median(BERTbase,axis=1)+ iqr(BERTbase,axis=1),
                 np.median(BERTbase,axis=1)- iqr(BERTbase,axis=1),
                 color='purple',alpha=.3)

for layer in range(BERTbase.shape[0]):
    plt.scatter(layer*np.ones(BERTbase.shape[1]),BERTbase[layer,:],color='purple',alpha=.1)

plt.grid(color = 'gray', linestyle = '--', linewidth = 0.5)
plt.title('BERT base')
plt.ylabel('scores')
plt.xlabel('layers')
plt.ylim(0,1)
plt.axhline(y=.5,color='k',linestyle='dashed')

plt.subplot(1,2,2)

plt.scatter(np.arange(GPT1_full.shape[0]),GPT1_full,color='black')

plt.plot(np.arange(GPT1.shape[0]),np.median(GPT1,axis=1),color='purple')
plt.fill_between(np.arange(GPT1.shape[0]),
                 np.median(GPT1,axis=1)+ iqr(GPT1,axis=1),
                 np.median(GPT1,axis=1)- iqr(GPT1,axis=1),
                 color='purple',alpha=.3)

for layer in range(GPT1.shape[0]):
    plt.scatter(layer*np.ones(GPT1.shape[1]),GPT1[layer,:],color='purple',alpha=.1)

plt.grid(color = 'gray', linestyle = '--', linewidth = 0.5)
plt.title('GPT-1')
plt.ylabel('scores')
plt.xlabel('layers')
plt.ylim(0,1)
plt.axhline(y=.5,color='k',linestyle='dashed')
```
