```python
import pickle
import numpy as np
from scipy.stats import sem
import matplotlib.pyplot as plt
```

```python
"""
Visualization of level of symmetry between different models 
(across layers and heads)
"""

plt.figure(figsize=(15,4))

## BERT models
plt.subplot(1,3,1)

# get scores for BERT models calculated on each head
with open('../_data/fig_scores/BERTmodels.pkl', 'rb') as file:
    models = pickle.load(file)

SList = [models[key][-2].flatten()**2 for key in list(models.keys())]
names = list(models.keys())

for i, model in enumerate(SList):
    
    plt.scatter(i*np.ones(SList[i].size),SList[i],color='purple',alpha=.01)
    plt.scatter(i,(SList[i].flatten()).mean(),color='purple')

# get scores for BERT models calculated on the whole M matrix
with open('../_data/fig_scores/BERTFullmodels.pkl', 'rb') as file:
    models = pickle.load(file)

SList = [models[key][-2].flatten()**2 for key in list(models.keys())]
names = list(models.keys())

for i, model in enumerate(SList):
    
    plt.scatter(i*np.ones(SList[i].size),SList[i],color='darkred',alpha=.01)
    plt.scatter(i,(SList[i].flatten()).mean(),color='darkred')

plt.grid(color = 'gray', linestyle = '--', linewidth = 0.5)
plt.title('BERT models')
plt.ylabel('scores')
plt.xticks(np.arange(len(names)),names)
plt.xticks(rotation=50)
plt.ylim(0,1)
plt.axhline(y=.5,color='k',linestyle='dashed')

## ROBERTA
plt.subplot(1,3,2)

with open('../_data/fig_scores/ROBERTAmodels.pkl', 'rb') as file:
    models = pickle.load(file)

SList = [models[key][-2].flatten()**2 for key in list(models.keys())]
names = list(models.keys())

for i, model in enumerate(SList):
    
    plt.scatter(i*np.ones(SList[i].size),SList[i],color='purple',alpha=.01)
    plt.scatter(i,(SList[i].flatten()).mean(),color='purple')

# get scores for BERT models calculated on the whole M matrix
with open('../_data/fig_scores/ROBERTAFullmodels.pkl', 'rb') as file:
    models = pickle.load(file)

SList = [models[key][-2].flatten()**2 for key in list(models.keys())]
names = list(models.keys())

for i, model in enumerate(SList):
    
    plt.scatter(i*np.ones(SList[i].size),SList[i],color='darkred',alpha=.01)
    plt.scatter(i,(SList[i].flatten()).mean(),color='darkred')

plt.grid(color = 'gray', linestyle = '--', linewidth = 0.5)
plt.title('ROBERTA models')
plt.ylabel('scores')
plt.xticks(np.arange(len(names)),names)
plt.xticks(rotation=50)
plt.ylim(0,1)
plt.axhline(y=.5,color='k',linestyle='dashed')

## ALBERT
plt.subplot(1,3,3)

with open('../_data/fig_scores/ALBERTmodels.pkl', 'rb') as file:
    models = pickle.load(file)

SList = [models[key][-2].flatten()**2 for key in list(models.keys())]
names = list(models.keys())

for i, model in enumerate(SList):
    
    plt.scatter(i*np.ones(SList[i].size),SList[i],color='purple',alpha=.01)
    plt.scatter(i,(SList[i].flatten()).mean(),color='purple')

# get scores for BERT models calculated on the whole M matrix
with open('../_data/fig_scores/ALBERTFullmodels.pkl', 'rb') as file:
    models = pickle.load(file)

SList = [models[key][-2].flatten()**2 for key in list(models.keys())]
names = list(models.keys())

for i, model in enumerate(SList):
    
    plt.scatter(i*np.ones(SList[i].size),SList[i],color='darkred',alpha=.01)
    plt.scatter(i,(SList[i].flatten()).mean(),color='darkred')

plt.grid(color = 'gray', linestyle = '--', linewidth = 0.5)
plt.title('ALBERT models')
plt.ylabel('scores')
plt.xticks(np.arange(len(names)),names)
plt.xticks(rotation=50)
plt.ylim(0,1)
plt.axhline(y=.5,color='k',linestyle='dashed')
```
