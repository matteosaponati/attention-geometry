```python
"""
mean symmetry scores (across layers and heads) as a function of total # parameters
"""

plt.figure(figsize=(10,4))

plt.subplot(1,2,1)

## encoder-only models
plt.scatter(BERT_par,BERT_score, color='darkblue', label='BERT')
plt.scatter(ROBERTA_par,ROBERTA_score,color='blue', label='ROBERTA')
plt.scatter(ALBERT_par,ALBERT_score,color='mediumslateblue', label='ALBERT', facecolors='none')
## encoder-decoder models
# plt.scatter(T5Encoder_par,T5Encoder_score,color='plum', label='T5 (encoder)')

plt.title('encoder-only models')
plt.axhline(y=.5,linestyle='dashed',color='k')
plt.grid(color = 'gray', linestyle = '--', linewidth = 0.5)
plt.xscale('log')
plt.legend()
plt.ylim(0,1)
plt.ylabel('symmetry score (mean)')
plt.xlabel('# parameters')
plt.xlim(1e5,10e10)

plt.subplot(1,2,2)

## decoder-only models
plt.scatter(GPT_par,GPT_score,color='palevioletred', label='GPT')
plt.scatter(OPT_par,OPT_score,color='darkmagenta', label='OPT')
plt.scatter(TinyGPT_par,TinyGPT_score,color='hotpink', label='TinyGPT')
## encoder-decoder models
# plt.scatter(T5Decoder_par,T5Decoder_score,color='plum', label='T5 (decoder)')
## decoder-only models (with quite some other fine-tuning)
plt.scatter(GEMMA1_par,GEMMA1_score,color='crimson', label='Gemma', facecolors='none')
plt.scatter(GEMMA_par,GEMMA_score,color='crimson', facecolors='none')
plt.scatter(LLAMA_par,LLAMA_score,color='magenta', label='Llama 2', facecolors='none')

plt.title('decoder-only models')
plt.axhline(y=.5,linestyle='dashed',color='k')
plt.grid(color = 'gray', linestyle = '--', linewidth = 0.5)
plt.xscale('log')
plt.legend()
plt.ylim(0,1)
plt.ylabel('symmetry score (mean)')
plt.xlabel('# parameters')
plt.xlim(1e5,10e10)
```
