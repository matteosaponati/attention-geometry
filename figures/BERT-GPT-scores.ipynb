```python
import pickle
import numpy as np
from scipy.stats import sem
import matplotlib.pyplot as plt
```

```python
"""
Visualization of level of symmetry between different models 
(across layers and heads)
"""

plt.figure(figsize=(8,8))

## BERT models
plt.subplot(2,2,1)

# get scores for BERT models calculated on each head
with open('../_data/fig_scores/BERT-query-key.pkl', 'rb') as file:
    models = pickle.load(file)

SList = [models[key][-2].flatten()**2 for key in list(models.keys())]
names = list(models.keys())

for i, model in enumerate(SList):
    
    plt.scatter(i*np.ones(SList[i].size),SList[i],color='purple',alpha=.01)
    plt.scatter(i,(SList[i].flatten()).mean(),color='purple')

# get scores for BERT models calculated on the whole M matrix
with open('../_data/fig_scores/BERT-query-key-full.pkl', 'rb') as file:
    models = pickle.load(file)

SList = [models[key][-2].flatten()**2 for key in list(models.keys())]
names = list(models.keys())

for i, model in enumerate(SList):
    
    plt.scatter(i*np.ones(SList[i].size),SList[i],color='darkred',alpha=.01)
    plt.scatter(i,(SList[i].flatten()).mean(),color='darkred')

plt.grid(color = 'gray', linestyle = '--', linewidth = 0.5)
plt.title('BERT models')
plt.ylabel('scores')
plt.xticks(np.arange(len(names)),[])
plt.ylim(0,1)
plt.axhline(y=.5,color='k',linestyle='dashed')

## GPT
plt.subplot(2,2,2)

with open('../_data/fig_scores/GPT-query-key.pkl', 'rb') as file:
    models = pickle.load(file)

SList = [models[key][-2].flatten()**2 for key in list(models.keys())]
names = list(models.keys())

for i, model in enumerate(SList):
    
    plt.scatter(i*np.ones(SList[i].size),SList[i],color='purple',alpha=.01)
    plt.scatter(i,(SList[i].flatten()).mean(),color='purple')

# get scores for BERT models calculated on the whole M matrix
with open('../_data/fig_scores/GPT-query-key-full.pkl', 'rb') as file:
    models = pickle.load(file)

SList = [models[key][-2].flatten()**2 for key in list(models.keys())]
names = list(models.keys())

for i, model in enumerate(SList):
    
    plt.scatter(i*np.ones(SList[i].size),SList[i],color='darkred',alpha=.01)
    plt.scatter(i,(SList[i].flatten()).mean(),color='darkred')

plt.grid(color = 'gray', linestyle = '--', linewidth = 0.5)
plt.title('GPT models')
plt.ylabel('scores')
plt.xticks(np.arange(len(names)),[])
plt.ylim(0,1)
plt.axhline(y=.5,color='k',linestyle='dashed')

## BERT models
plt.subplot(2,2,3)

# get scores for BERT models calculated on each head
with open('../_data/fig_scores/BERT-value-out.pkl', 'rb') as file:
    models = pickle.load(file)

SList = [models[key][-2].flatten()**2 for key in list(models.keys())]
names = list(models.keys())

for i, model in enumerate(SList):
    
    plt.scatter(i*np.ones(SList[i].size),SList[i],color='lightblue',alpha=.01)
    plt.scatter(i,(SList[i].flatten()).mean(),color='lightblue')

# get scores for BERT models calculated on the whole M matrix
with open('../_data/fig_scores/BERT-value-out-full.pkl', 'rb') as file:
    models = pickle.load(file)

SList = [models[key][-2].flatten()**2 for key in list(models.keys())]
names = list(models.keys())

for i, model in enumerate(SList):
    
    plt.scatter(i*np.ones(SList[i].size),SList[i],color='darkblue',alpha=.01)
    plt.scatter(i,(SList[i].flatten()).mean(),color='darkblue')

plt.grid(color = 'gray', linestyle = '--', linewidth = 0.5)
plt.ylabel('scores')
plt.xticks(np.arange(len(names)),names)
plt.xticks(rotation=50)
plt.ylim(0,1)
plt.axhline(y=.5,color='k',linestyle='dashed')

## GPT models
plt.subplot(2,2,4)

# get scores for BERT models calculated on each head
with open('../_data/fig_scores/GPT-value-out.pkl', 'rb') as file:
    models = pickle.load(file)

SList = [models[key][-2].flatten()**2 for key in list(models.keys())]
names = list(models.keys())

for i, model in enumerate(SList):
    
    plt.scatter(i*np.ones(SList[i].size),SList[i],color='lightblue',alpha=.01)
    plt.scatter(i,(SList[i].flatten()).mean(),color='lightblue')

# get scores for BERT models calculated on the whole M matrix
with open('../_data/fig_scores/GPT-value-out-full.pkl', 'rb') as file:
    models = pickle.load(file)

SList = [models[key][-2].flatten()**2 for key in list(models.keys())]
names = list(models.keys())

for i, model in enumerate(SList):
    
    plt.scatter(i*np.ones(SList[i].size),SList[i],color='darkblue',alpha=.01)
    plt.scatter(i,(SList[i].flatten()).mean(),color='darkblue')

plt.grid(color = 'gray', linestyle = '--', linewidth = 0.5)
plt.ylabel('scores')
plt.xticks(np.arange(len(names)),names)
plt.xticks(rotation=50)
plt.ylim(0,1)
plt.axhline(y=.5,color='k',linestyle='dashed')
```
